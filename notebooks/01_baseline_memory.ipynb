# Notebook 1: Baseline Memory Profiling
Measures baseline GPU memory usage for GPT-2 medium in FP16.

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

MODEL_NAME = "gpt2-medium"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForCausalLM.from_pretrained(MODEL_NAME, dtype=torch.float16).cuda()

prompt = "The memory bottleneck in large language models"
inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

torch.cuda.reset_peak_memory_stats()
with torch.no_grad():
    out = model.generate(**inputs, max_new_tokens=50)

peak_mem = torch.cuda.max_memory_allocated() / 1e6
print(f"Peak Memory: {peak_mem:.2f} MB")
print(tokenizer.decode(out[0]))
```
